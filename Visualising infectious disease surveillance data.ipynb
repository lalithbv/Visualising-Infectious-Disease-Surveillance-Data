````python
# Function to infer freq of datetime index of the dataframe

def add_freq(idx, freq=None):
    """Add a frequency attribute to idx, through inference or directly.

    Returns a copy.  If `freq` is None, it is inferred.
    """

    idx = idx.copy()
    if freq is None:
        if idx.freq is None:
            freq = pd.infer_freq(idx)
        else:
            return idx
    idx.freq = pd.tseries.frequencies.to_offset(freq)
    if idx.freq is None:
        raise AttributeError('no discernible frequency found to `idx`.  Specify'
                             ' a frequency string with `freq`.')
    return idx
````

### Forecasting with the Holt-Winters Method

### <font color=blue>Simple Exponential Smoothing / Simple Moving Average</font>
This is the simplest to forecast. $\hat{y}$ is equal to the most recent value in the dataset, and the forecast plot is simply a horizontal line extending from the most recent value.
### <font color=blue>Double Exponential Smoothing / Holt's Method</font>
This model takes trend into account. Here the forecast plot is still a straight line extending from the most recent value, but it has slope.
### <font color=blue>Triple Exponential Smoothing / Holt-Winters Method</font>
This model has (so far) the "best" looking forecast plot, as it takes seasonality into account. When we expect regular fluctuations in the future, this model attempts to map the seasonal behavior.

````python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
````

````python
import warnings
warnings.filterwarnings("ignore")

model_hw = ExponentialSmoothing(influenza_cases_train['Cases'],trend='mul',seasonal='mul',seasonal_periods=52).fit()
````

````python
predictions_hw = model_hw.forecast(52).rename('HW Forecast')
````

### ACF Plots

Plotting the magnitude of the autocorrelations over the first four year lag.

**Autocorrelation for 1D:**

The correlation $\rho$ (rho) between two variables $y_1,y_2$ is given as:

$\rho = \frac {\operatorname E[(y_1âˆ’\mu_1)(y_2âˆ’\mu_2)]} {\sigma_{1}\sigma_{2}} = \frac {\operatorname {Cov} (y_1,y_2)} {\sigma_{1}\sigma_{2}}$,

where $E$ is the expectation operator, $\mu_{1},\sigma_{1}$ and $\mu_{2},\sigma_{2}$ are the means and standard deviations of $y_1$ and $y_2$.

When working with a single variable (i.e. <em>autocorrelation</em>) we would consider $y_1$ to be the original series and $y_2$ a lagged version of it. Note that with autocorrelation we work with $\bar y$, that is, the full population mean, and <em>not</em> the means of the reduced set of lagged factors (see note below).

Thus, the formula for $\rho_k$ for a time series at lag $k$ is:

${\displaystyle \rho_k = \frac {\sum\limits_{t=1}^{n-k} (y_t - \bar{y})(y_{t+k}-\bar{y})} {\sum\limits_{t=1}^{n} (y_t - \bar{y})^2}}$

This can be written in terms of the covariance constant $\gamma_k$ as:

${\displaystyle \rho_k = \frac {\gamma_k n} {\gamma_0 n} = \frac {\gamma_k} {\sigma^2}}$

Note that ACF values are bound by -1 and 1. That is, ${\displaystyle -1 \leq \rho_k \leq 1}$

````python
from statsmodels.graphics.tsaplots import plot_acf,plot_pacf
````

````python
title = 'Autocorrelation of weekly influenza cases'
# lags = len(influenza_cases)-1
lags = 52*4
plot_acf(influenza_cases,title=title,lags=lags);
plt.xticks(np.arange(0, lags+1, 52));
plt.xlabel("Weekly lag");
plt.ylabel("Correlation");
````

**Shaded region is 95% CI. Correlation values not in the shaded region implies that they are very likely to be a correlation.**

## SEIR Model

**Incubation Period** is the period between exposure to an infection and the appearance of the first symptoms.

Flu symptoms usually appear 1 to 3 days after becoming infected. People are potentially infectious to others even before they have symptoms.

Ref: http://conditions.health.qld.gov.au/HealthCondition/condition/14/217/82/influenza-the-flu

**Infectious Period** is the time interval during which a host is infectious, i.e. capable of directly or indirectly transmitting pathogenic infectious agents or pathogens to another susceptible host.

Patients may shed influenza virus for up to 24 hours (1 day) before onset of symptoms and until 7 days after the onset of symptoms. Viral shedding in adults peaks in the first 1 to 2 days after symptom onset, then reduces to very low levels by 5 days after onset of symptoms. Not all cases of influenza infection exhibit fever, but when it is present, it is correlated with viral shedding.

Children and younger adults may shed influenza virus for 10 or more days, and immunosuppressed persons may shed virus for weeks. However, the ability to transmit infection is likely to be higher when respiratory symptoms are present.

Patients are considered no longer infectious if 24 hours have elapsed since the resolution of fever, provided either:

have received 72 hours of anti-influenza medication or
5 days have elapsed since onset of respiratory symptoms.

Ref: https://www.health.nsw.gov.au/Infectious/controlguideline/Pages/influenza.aspx


The **transmissibility** of the virus between humans will affect the breadth and speed of spread across the globe
and the Australian community. The transmissibility of a pandemic influenza virus is usually estimated to have a
basic **reproduction number (R0) of 1.2â€“2.5**. For communicable diseases, this is not particularly high. However,
the serial interval for influenza (the time taken between generations of infection) is short (around two days), with
transmission from a primary case potentially occurring before symptom onset, allowing infection to spread
relatively quickly. 

Ref: https://www1.health.gov.au/internet/main/publishing.nsf/content/519F9392797E2DDCCA257D47001B9948/$File/w-AHMPPI-2019.PDF

Australiaâ€™s population was 25,739,256 people at 30 June 2021.

Ref: https://www.abs.gov.au/statistics/people/population/national-state-and-territory-population/latest-release

In 2017â€“18, Australia had **3.9 beds per 1,000 population** in public and private hospitals.

Ref: https://www.aihw.gov.au/reports/hospitals/hospital-resources-2017-18-ahs/contents/hospitals-and-average-available-beds

````python
from scipy.integrate import odeint
````

````python
def SEIR(u, title):

    t_incubation = 2 # Assuming incubation period = 2 days
    t_infective = 3 # Assuming infectious period = 3 days
    R0 = 1.85 # Assuming R0 = (1.2+2.5) / 2
    N = 25739256 # Australian Population
    percent_pop_of_beds = round((3.9/1000) * 100, 2)
    
    # initial number of infected and recovered individuals
    e0 = 1/N
    i0 = 0.00
    r0 = 0.00
    s0 = 1 - e0 - i0 - r0
    x0 = [s0,e0,i0,r0]

    alpha = 1/t_incubation
    gamma = 1/t_infective
    beta = R0*gamma
    
    def epidemic(x,t):
        s,e,i,r = x
        dx = np.zeros(4)
        dx[0] = -(1-u)*beta * s * i
        dx[1] = (1-u)*beta * s * i - alpha * e
        dx[2] = alpha * e - gamma * i
        dx[3] = gamma*i
        return dx

    t = np.arange(0, 366, 2)
    x = odeint(epidemic,x0,t)
    s = x[:,0]; e = x[:,1]; i = x[:,2]; r = x[:,3]

    # plot the data
    plt.figure(figsize=(18,10), dpi=500)

    plt.subplot(3,1,1)
    plt.title(f'{title} (u = {str(u*100)}%)')
    plt.plot(t,s, color='blue', lw=3, label='Susceptible')
    plt.plot(t,r, color='red',  lw=3, label='Recovered')
    plt.ylabel('Fraction')
    plt.legend()
    
    plt.subplot(3,1,2)
    plt.plot(t,i, color='orange', lw=3, label='Infective')
    plt.plot(t,e, color='purple', lw=3, label='Exposed')
    plt.axhline(y=percent_pop_of_beds/ 100, linestyle='--', label='Healthcare System Capacity (0.39% of population)')
    plt.ylim(0, 0.08)
    plt.ylabel('Fraction')
    plt.legend()
    
    plt.subplot(3,1,3)
    plt.plot(t,i*0.0714, color='green', lw=3, label='Hospitalized (7.14% of infective)') # Assuming 7.14% hospitalization rate
    plt.axhline(y=percent_pop_of_beds/ 100, linestyle='--', label='Healthcare System Capacity (0.39% of population)')
    plt.ylim(0, 0.006)
    plt.xlabel('Time (days)')
    plt.ylabel('Fraction')
    plt.legend()

    plt.show();
````

````python
scaled_train = scaler.transform(influenza_cases_train['Cases'].values.reshape(-1, 1))
scaled_test = scaler.transform(influenza_cases_test['Cases'].values.reshape(-1, 1))
````

````python
# Let's redefine to get 52 weeks back and then predict the next week out
n_input = 52
n_features = 1
generator = TimeseriesGenerator(scaled_train, scaled_train, length=n_input, batch_size=2)
````

````python
# What does the first batch look like?
X,y = generator[0]
````

````python
y
````

````python
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
````

````python
model_LSTM = Sequential()
model_LSTM.add(LSTM(200, activation='relu', input_shape=(n_input, n_features)))
model_LSTM.add(Dense(1))
model_LSTM.compile(optimizer='adam', loss='mse')
````

````python
model_LSTM.summary()
````

````python
# fit model
model_LSTM.fit_generator(generator,epochs=50)
````

````python
loss_per_epoch = model_LSTM.history.history['loss']
plt.plot(range(len(loss_per_epoch)),loss_per_epoch);
````

````python
predictions_LSTM = []

first_eval_batch = scaled_train[-n_input:]
current_batch = first_eval_batch.reshape((1, n_input, n_features))

for i in range(len(influenza_cases_test)):
    
    # get prediction 1 time stamp ahead ([0] is for grabbing just the number instead of [array])
    current_pred = model_LSTM.predict(current_batch)[0]
    
    # store prediction
    predictions_LSTM.append(current_pred) 
    
    # update batch to now include prediction and drop first value
    current_batch = np.append(current_batch[:,1:,:],[[current_pred]],axis=1)
````

````python
true_predictions_LSTM = scaler.inverse_transform(predictions_LSTM)
````

````python
influenza_cases_test_pred = influenza_cases_test.copy()
````

````python
influenza_cases_test_pred['Predictions'] = true_predictions_LSTM
````
